```python
returnBody = ''

myKey = 'A141'
myColumn = df.loc[:, myKey]
```

```python
merged = pd.merge(df,df2,left_on='A141', right_on='Type of Service', how='inner')
```

```python
# select multiple columns
selected = merged[['A141','Type of Service', 'ID_Type of Service']]
selected.to_json(orient='records')
```

```python
for items in myColumn.iteritems():
    row = df2.loc[df2['Type of Service'] == items[1]]
    # print (row)
    if not(row.empty):  
        # print(row.iloc[0]['Type of Service'], row.iloc[0]['ID_Type of Service']);
        returnBody = returnBody + '<Option value="{}">{}</Option>\n'.format(row.iloc[0]['ID_Type of Service'],row.iloc[0]['Type of Service'])

print(returnBody)
```
